```python
# pie chart >> used to understaand the persentage of the data 
langs = ["C", "Python","Java"]
std = [20,100,40]

plt.pie(std,labels = langs , autopct= "%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
# focus on one segment 
explode = (0.0,0.1,0.1)
langs = ["C", "Python","Java"]
std = [20,100,40]

plt.pie(std,labels = langs , autopct= "%1.1f%%", explode = explode)
plt.title("Pie chart")
plt.show()
```

```python
# focus on one segment 
explode = (0.0,0.1,0.1)
langs = ["C", "Python","Java"]
std = [20,100,40]

plt.pie(std,labels = langs , autopct= "%1.1f%%", explode = explode, shadow=True)
plt.title("Pie chart")
plt.show()
```

```python
# area plot  >> variables changed over time

years = [2021,2018,2022,2023,2017]
category1 = [20,25,30,28,35]
category2 = [10,15,20,18,25]
category3 = [5,8,10,9,15]
plt.stackplot(years,category1,category2,category3,labels=["category1","category2","category3"],
              colors=["skyblue","lightgreen","orange"])
plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
