```python
#MERGE 
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({ "key1" :[1,2,4,5,6],
                     "key2" : [4,5,6,7,8],
                     "key3" : [3,4,5,6,6]
                     })
df1
```

```python
df2 = pd.DataFrame({ "key1" : [1,2,45,6,67],
                     "key4" : [56,5,6,7,8],
                     "key5"  : [3,56,5,6,6]
})
df2
```

```python
# inner merge
pd.merge(df1,df2,how='inner')
```

```python
# left merge
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how = 'right')
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,how='outer') # all the comon element of the common element from bot DF
```

```python
pd.merge(df1,df2,how='cross')
```
